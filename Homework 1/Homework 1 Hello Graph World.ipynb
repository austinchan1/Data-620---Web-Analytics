## Import Necessary Packages

The code below imports the necessary packages for the assignment. The first package I imported is the networkx package. This package is useful for creating and manipulating graph objects. The second package I imported is matplotlib. This package is useful for showing graphs visually as images.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Creating The Graph

The code below creates the graph specified in the assignment.

The first line creates the graph object and names it "G".

The second line adds the edges of the graph by specifying the node pairs 
for each individual edge.

The third and fourth line draws and displays the graph accordingly.

```python
G = nx.Graph()


G.add_edges_from([("Andre","Carol"),
                  ("Andre","Diane"),
                  ("Andre","Beverly"),
                  ("Andre","Fernando"),
                  ("Beverly","Diane"),
                  ("Beverly","Ed"),
                  ("Beverly","Garth"),
                  ("Carol","Diane"),
                  ("Carol","Fernando"),
                  ("Diane","Ed"),
                  ("Diane","Fernando"),
                  ("Diane","Garth"),
                  ("Ed","Garth"),
                  ("Fernando","Garth"),
                  ("Fernando","Heather"),
                  ("Garth","Heather"),
                  ("Heather","Ike"),
                  ("Ike","Jane")])

nx.draw(G, with_labels = True, font_weight="bold")
plt.show()
```
